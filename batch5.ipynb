```python
import pandas as pd
```

### Check Products

```python
product_staging = pd.read_excel("check-batch-5/Product_staging.xlsx")
product_production = pd.read_excel("check-batch-5/Product_production.xlsx")

product_staging["ID"] = product_staging["ID"].astype(str).str.strip().str.lower()
product_production["ID"] = product_production["ID"].astype(str).str.strip().str.lower()

set_staging = set(product_staging["ID"])
set_production = set(product_production["ID"])
```

```python
print(f"Total Staging : {product_staging.shape[0]}")
print(f"Total Production : {product_production.shape[0]}")
```

```python
## Show difference data production on staging
# print("Production vs Staging: {}".format(len(set_production - set_staging)))
# for i in list(set_production - set_staging):
#     print(i)
```

```python
# compare different rows in two files (staging and production)
diff_rows_product_staging = pd.merge(product_staging, product_production, how='outer', indicator=True).loc[lambda x: x.pop('_merge').eq('left_only')]
diff_rows_product_production = pd.merge(product_production, product_staging, how='outer', indicator=True).loc[lambda x: x.pop('_merge').eq('left_only')]

print(f"Total Different Rows Product in Staging : {diff_rows_product_staging.shape[0]}")
print(f"Total Different Rows Product in Production : {diff_rows_product_production.shape[0]}")
```

### Check Product Variant (Product Stock)

```python
product_stock_staging = pd.read_excel("check-batch-5/Product_variant_staging.xlsx")
product_stock_production = pd.read_excel("check-batch-5/Product_variant_production.xlsx")

product_stock_staging["Display Name"] = product_stock_staging["Display Name"].astype(str).str.strip().str.lower()
product_stock_production["Display Name"] = product_stock_production["Display Name"].astype(str).str.strip().str.lower()

set_staging = set(product_stock_staging["Display Name"])
set_production = set(product_stock_production["Display Name"])
```

```python
print(f"Total Staging : {product_stock_staging.shape[0]}")
print(f"Total Production : {product_stock_production.shape[0]}")
```

```python
# check difference Average Cost, Free To Use Quantity, Incoming, Outgoing, Quantity On Hand, Total Value
print(f"Average Cost staging: {product_stock_staging['Average Cost'].sum()}")
print(f"Average Cost production: {product_stock_production['Average Cost'].sum()}")
print(f"Different Average Cost: {product_stock_staging['Average Cost'].sum() - product_stock_production['Average Cost'].sum()}")
print("-" * 50)
print(f"Free To Use Quantity staging: {product_stock_staging['Free To Use Quantity'].sum()}")
print(f"Free To Use Quantity production: {product_stock_production['Free To Use Quantity'].sum()}")
print(f"Different Free To Use Quantity: {product_stock_staging['Free To Use Quantity'].sum() - product_stock_production['Free To Use Quantity'].sum()}")
print("-" * 50)
print(f"Incoming staging: {product_stock_staging['Incoming'].sum()}")
print(f"Incoming production: {product_stock_production['Incoming'].sum()}")
print(f"Different Incoming: {product_stock_staging['Incoming'].sum() - product_stock_production['Incoming'].sum()}")
print("-" * 50)
print(f"Outgoing staging: {product_stock_staging['Outgoing'].sum()}")
print(f"Outgoing production: {product_stock_production['Outgoing'].sum()}")
print(f"Different Outgoing: {product_stock_staging['Outgoing'].sum() - product_stock_production['Outgoing'].sum()}")
print("-" * 50)
print(f"Quantity On Hand staging: {product_stock_staging['Quantity On Hand'].sum()}")
print(f"Quantity On Hand production: {product_stock_production['Quantity On Hand'].sum()}")
print(f"Different Quantity On Hand: {product_stock_staging['Quantity On Hand'].sum() - product_stock_production['Incoming'].sum()}")
print("-" * 50)
print(f"Total Value staging: {product_stock_staging['Total Value'].sum()}")
print(f"Total Value production: {product_stock_production['Total Value'].sum()}")
print(f"Different Total Value: {product_stock_staging['Total Value'].sum() - product_stock_production['Total Value'].sum()}")
```

```python
# compare different rows in two files (staging and production)
diff_rows_staging = pd.merge(product_stock_staging, product_stock_production, how='outer', indicator=True).loc[lambda x: x.pop('_merge').eq('left_only')]
diff_rows_production = pd.merge(product_stock_production, product_stock_staging, how='outer', indicator=True).loc[lambda x: x.pop('_merge').eq('left_only')]

print(f"Total Different Rows Staging : {diff_rows_staging.shape[0]}")
print(f"Total Different Rows Production : {diff_rows_production.shape[0]}")
```
