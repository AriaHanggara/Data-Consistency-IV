```python
import pandas as pd
```

### Check Products

```python
product_staging = pd.read_excel("check-batch-4/Product_staging.xlsx")
product_production = pd.read_excel("check-batch-4/Product_production.xlsx")

product_staging["ID"] = product_staging["ID"].astype(str).str.strip().str.lower()
product_production["ID"] = product_production["ID"].astype(str).str.strip().str.lower()

set_staging = set(product_staging["ID"])
set_production = set(product_production["ID"])
```

```python
print(f"Total Staging : {product_staging.shape[0]}")
print(f"Total Production : {product_production.shape[0]}")
```

```python
product_production[product_production['Barcode'].duplicated()]
```

```python
product_staging['Barcode'].duplicated().sum()
```

```python
product_staging[product_staging['Barcode'].duplicated()]
```

```python
print("Staging vs Production: {}".format(len(set_staging - set_production)))
for i in list(set_staging - set_production):
    print(i)
```

```python
print("Production vs Staging: {}".format(len(set_production - set_staging)))
for i in list(set_production - set_staging):
    print(i)
```

### Check Product Packaging
